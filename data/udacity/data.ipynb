```python
import matplotlib.pyplot as plt
import matplotlib.cbook as cbook

import numpy as np
import pandas as pd

plt.rcParams["figure.figsize"] = (25,6)

data = pd.read_csv('Y_train.csv')

fig,ax=plt.subplots()
ax.plot(data.index, data.angle, color="red")
ax.set_xlabel("index")
ax.set_ylabel("angle")

ax2=ax.twinx()
ax2.plot(data.index, data.speed, color="blue")
ax2.set_ylabel("speed")
```
